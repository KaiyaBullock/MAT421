# 3.2 Continuity and Differentiation

A function $f(x)$ is **continuous** at $a$ if the limit of $f(x)$ as $x$ approaches $a$ is $f(a)$, meaning that the function is smooth with no abrupt value changes.

This is demonstarted in the graph below with the continuous function $y=2x$ in red and the non-continuous function $y=1/x$ in blue

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-4,4,50)
y1=2*x
y2=1/x

plt.figure
plt.plot(x,y1,'-r')
plt.plot(x,y2,'-b')
plt.title('Continuity')
```

### Extreme Value Theorem
Let $f:D→R$ be a real-valued, continuous function on a nonempty, closed, bounded set $D⊆R^d$. Then $f$ attains a maximum and minimum on D.

### Derivatives
Let $f:D→R$ where $D⊆R$ and let $x_0∈D$ be an interior point of D. The derivative of $f$ at $x_0$ is $f'(x_0)=\frac{df(x_0)}{dx}=lim_{h→0}\frac{f(x_0+h)-f(x_0)}{h}$ provided the limit exists.

The derivative of a function measures the sensitivity to change of the output with respect to another variable.

```python
from sympy import *
import numpy as np

x=Symbol('x')
y=x**(4)
yprime=y.diff(x)
yprime
```

### Rolle Theorem
Let $f:[a,b]→R$ be a continuous function and assume that its derivative exists on $(a,b)$. If $f(a)=f(b)$, then there is $a<c<b$ such that $f'(c)=0$
### Mean Value Theorem
Let $f:[a,b]→R$ be a continuous function and assume that its derivative exists on $(a,b)$. If $f(a)=f(b)$, then there is $a<c<b$ such that $f(b)=f(a)+(b-a)f'(c)$

A **partial derivative** is when you take a derivative with respect to one variable while holding the other variables constant. There is an example below

```python
from sympy import symbols, diff

x,y,z=symbols('x y z', real=True)
f = 10*x*y*z + x*cos(z) + y**2 + z**6*y

partial_x=diff(f,x)
partial_y=diff(f,y)
partial_z=diff(f,z)

print("partial_x=", partial_x)
print("partial_y=", partial_y)
print("partial_z=", partial_z)
```

### Taylor's Theorem
This theorem gives an approxiamtion of a differentiable function arouond a given point by a polynomial

Let $f : D → R$ where $D⊆R$. Suppose $f$ has a $m$ times continuous derivative of $[a,b]$. Then $f(b) = f(a) + (b−a)f′(a) +\frac{1}{2}(b−a)^2 f′′(a)+...+\frac{((b−a)^{m−1}}{(m−1)!} f^(m−1)(a)+Rm$. Where $Rm = \frac{(b-a)^m}{m!}f^m (a+θ(b-a))$ for some $0 < θ < 1$.


```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-np.pi, np.pi, 200)
y = np.zeros(len(x))

labels=['First Order', 'Third Order']

plt.figure(figsize=(10,8))
for n, label in zip(range(2), labels):
  y = y + ((-1)**n * x**(2*n+1)) / np.math.factorial(2*n+1)
  plt.plot(x, y, label = label)

plt.plot(x, np.sin(x), 'k', label = 'Analytic')
plt.grid()
plt.title('Taylor Series Approx. of Various Orders')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

# 3.3 Unconstrained Optimization
Let $f:R^d→R$. The point $x^{*}∈R^d$ is a **global minimizer** of $f$ over $R^d$ if $f(x)≥f(x^{∗}), ∀x∈R^d$.

Let $f:R^d→R$. The point $x^∗∈R^d$ is a **local minimizer** of $f$ over $R^d$ if there is $δ>0$ such that $f(x)≥f(x^∗), ∀x∈B_{δ}(x^∗)$ \ {$x^∗$}.

Let $f:R^d→R$. A vector $v$ is a **descent direction** for $f$ at $x_0$ if there is $α^∗>0$ such that $f(x_0 +αv)<f(x_0), ∀α∈(0,α^∗ )$.

**First-Order Necessary Condition:** Let $f:R^d→R$ be continuously differentiable on $R^d$. If $x_0$ is a local minimizer, then $∇f(x_0)=0$

**Second-Order Necessary Condition:** Let $f:R^d→R$ be twice continuously differentiable on $R^d$. If $x_0$ is a local minimizer, then $H_f(x_0)$ is PSD (positive semi-definite).

**Convex Set:** A set $D⊆R^d$ is convex if for all $x,y∈D$ and all $α∈[0,1]$,
  $(1− α)x+αy∈D$.

**Convex Function:** A function $f:R^d→R$ is convex if, for all $x,y∈R^d$ and all $α∈[0,1]$,
  $f((1− α)x+αy)≤(1− α)f(x)+αf(y)$.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,8)
y = x**2

x1 = np.linspace(0,6)
y1 = x*4.5

plt.plot(x, y, 'b', x1, y1, 'r')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
