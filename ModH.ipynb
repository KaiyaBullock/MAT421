# 22.1 ODE Ordinary Differential Equation
A differential equation is a relationship between a function $f(x)$, its independent variable $x$, and any number of its derivtaives. An ODE is a differential equation where the independent variable, and the derivatives, is in one dimension.

nth order ODE: $F(x,f(x),\frac{df(x)}{dx},\frac{d^2f(x)}{dx^2},...,\frac{d^{n-1}f(x)}{dx^{n-1}})=\frac{d^nf(x)}{dx^n}$

For an ODE of order $n$ a particular situation is a $p(x)$ that satisfies the differential equation and *n* explicitly known values of the solution, or its derivatives, at certain points.

Generally stated $p(x)$ must satisfy the differential equation and $p^{(j)}(x_i)=p_i$ where $p^{(j)}$ is the $j^{th}$ derivative of $p$, for *n* triplets, $(j,x_i,p_i)$

# 22.2 Reduction of Order
Many numerical methods for solving initial value problems are designed specifically to solve first-order differential equations. To make these solvers useful fro solving higher order differential equations, we must often reduce the order.

$f^{(n)}(t)=F(t,f(t),f^{(1)},...,f^{(n-1)}(t))$

# 22.3 The Euler Method


```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline

f = lambda t, s: np.exp(-t)
h = 0.01
t = np.arange(0, 1+h, h)
s0 = -1

s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t)-1):
  s[i+1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12,8))
plt.plot(t, s, 'b--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution\
  for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
