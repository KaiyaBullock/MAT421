# Interpolation Problem Statement

Given a set of data with independent values ($x$) and dependent values ($y$), interpolation allows us to estimate a function ($\hat{y}(x)=y$) that will correspond these values without having to model.

# Linear Interpolation

In linear interpolation, an estimated point is assumed to lie on the line connecting its nearest points to the left and right.

The linear interpolatiion at x is:
$\hat{y}(x)=y_{i}+\frac{(y_{i+1}-y_{i})(x-x_{i})}{(x_{i+1}-x_{i})}$

Example: Find the linear interpolation at $x=0.5$ based on the data x = [0,1,2], y = [3,1,2]

The interpolation function yeilds:
$\hat{y}(x)=3+\frac{(1-3)(0.5-0)}{(1-0)}=2.0$

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
```

```python
x = [0, 1, 2]
y = [3, 1, 2]

f = interp1d(x,y)
y_hat = f(0.5)
print(y_hat)
```

```python
plt.figure(figsize = (10, 8))
plt.plot(x, y, '-ob')
plt.plot(0.5, y_hat, 'ro')
plt.title('Linear Interpolation at x = 0.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Cubic Sline Interpolation

In cublic slpine interpolation, the interpolating functiion is a set ofd of points being connected by a line, the are joined by the cubic polynomial $S_{i}(x)=a_{i}x^{3}+b_{i}x^{2}+c_{i}x+d_{i}$

For $n$ number of points there are $n-1$ cubic functions and each has 4 coefficients that we must find.

  Example: Plot the cubic slpine interpolation of the data set x = [0, 1, 2], y = [3, 1, 2] for $0 \le x \le 2$

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

x = [0, 1, 2]
y = [3, 1, 2]

f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 2, 100)
y_new = f(x_new)

plt.figure(figsize = (10, 8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
